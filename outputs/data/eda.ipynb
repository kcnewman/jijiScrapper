```python
def extract_all_properties(df: pd.DataFrame, drop: bool = True) -> pd.DataFrame:
    """Expand the property column into separate columns

    Args:
        df (pd.DataFrame): Input Dataframe with the 'properties' column
        drop (bool): Drop properties column in input Dataframe

    Returns:
        pd.DataFrame: Output Dataframe with the new columns
    """
    df["properties"] = df["properties"].apply(ast.literal_eval)
    properties = df["properties"].apply(pd.Series)
    df_out = df.join(properties)
    if drop:
        df_out = df_out.drop("properties", axis=1)

    return df_out
```

```python
df = extract_all_properties(df)
```

```python
def extract_all_ameneties(df):
    amenities_encoded = df["amenities"].str.strip().str.get_dummies(sep=",")
    df = df.join(amenities_encoded).drop("amenities", axis=1)

    return df
```

```python
df = extract_all_ameneties(df)
```
