```python
df["sub_loc"] = df["location"].str.split(",").str.get(1).str.strip()
df[["sub_loc", "location"]].head()
```

```python
df["sub_loc"].unique()
```

```python
def getSubLoc(loc):
    parts = str(loc).split(",")
    if len(parts) >= 3:
        return parts[1].strip()
    return parts[0]
```

```python
df["sub_loc"] = df["location"].apply(getSubLoc)
```
